```python
# Import Pandas.
import pandas as pd

# Import the CSV files.
classlist_a = pd.read_csv('classlist_a.csv')
classlist_b = pd.read_csv('classlist_b.csv')
classlist_c = pd.read_csv('classlist_c.csv')
classlist_d = pd.read_csv('classlist_d.csv')

# Concatenate classlists a and b; and c and d.
classlist_ab = pd.concat([classlist_a, classlist_b])
classlist_cd = pd.concat([classlist_c, classlist_d])

# View DataFrames.
print(classlist_ab.shape)
print(classlist_cd.shape)
```

```python
# Question

"""You decide to use the merge() function to establish whether there are indeed 675 students. To save time, you decide to merge classlist_ab and classlist_cd. Prepare your code snippet as follows:

*Create a new DataFrame with the name classlist_final.
*Use the pd.merge() function. 
*The tables to merge are classlist_ab (left) and classlist_cd (right).
*The parameter how='outer' indicates that all the data points/values will be included.
*The on='Student number' parameter indicates the key on which the joining of the tables will be performed."""


# Merge class lists ab and cd.
classlist_final = pd.merge(classlist_ab, classlist_cd, 
                           how='outer', on='Student number') 

# View the DataFrame.
print(classlist_final.shape)
classlist_final
```
